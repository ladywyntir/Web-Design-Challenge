# CSV to HTML
Using Pandas to import a CSV file to a DataFrame, then print that out as an HTML file.

```python
# import dependencies
import os
import pandas as pd

# load the file into a new DataFrame
csv_DF = pd.read_csv("weatherpy_output.csv")

csv_DF.head()
```

```python
csv_to_html = csv_DF.to_html('tabledata.html', index=False, classes=['table', 'table-striped', 'table-hover'])
```
